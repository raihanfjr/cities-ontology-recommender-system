```python
import pandas as pd
import numpy as np
```

```python
from owlready2 import *
onto_path.append("/path/to/your/local/ontology/repository")
onto = get_ontology("https://files.ifi.uzh.ch/ddis/oldweb/ddis/fileadmin/ont/nli/geography.owl")
onto.load()
```

```python
data = {'name': [],
        'area': [],
        'population': [],
        'pop_density': [],
        'lowest_elevation': [],
        'highest_elevation': []}

df_states = pd.DataFrame(data)

states_data = onto.search(type = onto.State)
low_elevation_data = onto.search(type = onto.LoPoint)
high_elevation_data = onto.search(type = onto.HiPoint)

for state in states_data:
    statename = state.name
    for prop in state.get_properties():
        for value in prop[state]:
            if prop.python_name == "statePopulation":
                statearea = int(value)
            elif prop.python_name == "stateArea":
                statepop = int(value)
            
            for low_elevation in  low_elevation_data:
                if(low_elevation.isLowestPointOf[0].name == statename):
                    stateloel = int(low_elevation.loElevation[0])
            
            for high_elevation in  high_elevation_data:
                if(high_elevation.isHighestPointOf[0].name == statename):
                    statehiel = int(high_elevation.hiElevation[0])
    
    statepopdense = statepop / statearea
    state_data = {'name': statename, 'area': statearea, 'population': statepop, 'pop_density': statepopdense, 'lowest_elevation': stateloel, 'highest_elevation': statehiel}
    df_states = df_states.append(state_data, ignore_index=True)

df_states = df_states.sort_values('pop_density')

display(df_states)  
```

```python
data = {'name': [],
        'population': [],
        'state': [],
        'capital': []}

df_cities = pd.DataFrame(data)

cities_data = onto.search(type = onto.City)
capital_data = onto.search(type = onto.Capital)

for city in cities_data:
    cityname = city.name
    capital = False
    for prop in city.get_properties():
        for value in prop[city]:
            if prop.python_name == "cityPopulation":
                citypop = int(value)
            elif prop.python_name == "isCityOf":
                citystate = value.name
    
    city_data = {'name': cityname, 'population': citypop, 'state': citystate, 'capital': capital}
    df_cities = df_cities.append(city_data, ignore_index=True)

for city in capital_data:
    cityname = city.name
    capital = True
    for prop in city.get_properties():
        for value in prop[city]:
            if prop.python_name == "cityPopulation":
                citypop = int(value)
            elif prop.python_name == "isCityOf":
                citystate = value.name
    
    city_data = {'name': cityname, 'population': citypop, 'state': citystate, 'capital': capital}
    df_cities = df_cities.append(city_data, ignore_index=True)    

df_cities = df_cities.sort_values('population')

print(df_cities)    
```

```python
stats = df_cities.describe()

# print(stats)
pop_min = stats.loc['min', 'population']
pop_med = stats.loc['50%', 'population']
pop_max = stats.loc['max', 'population']

print(pop_min)
print(pop_med)
print(pop_max)
```

```python
def rateState(city, stateDense):      
    doms = 0
    city_popdense = int(city.isCityOf[0].stateArea[0]) / int(city.isCityOf[0].statePopulation[0])
    if(stateDense == "Rural" or stateDense == "rural"):
        doms = (pd_med - city_popdense) / (pd_med - pd_min)
        
    elif(stateDense == "Suburban" or stateDense == "suburban"):
        if(city_popdense <= pd_med):
            doms = (city_popdense - pd_min) / (pd_med - pd_min)
        elif(city_popdense > pd_med):
            doms = (pd_max - city_popdense) / (pd_max - pd_med)
            
    elif(stateDense == "Urban" or stateDense == "urban"):
        doms = (city_popdense - pd_med) / (pd_max - pd_med)
    
    if(doms < 0):
        doms = 0
    
    return doms
```

```python
def rateCity(city, cityPop):      
    domc = 0
    city_pop = int(city.cityPopulation[0])
    if(cityPop == "Small" or cityPop == "small"):
        domc = (pop_med - city_pop) / (pop_med - pop_min)
        
    elif(cityPop == "Medium" or cityPop == "medium"):
        if(city_pop <= pop_med):
            domc = (city_pop - pop_min) / (pop_med - pop_min)
        elif(city_pop > pop_med):
            domc = (pop_max - city_pop) / (pop_max - pop_med)
            
    elif(cityPop == "Large" or cityPop == "large"):
        domc = (city_pop - pop_med) / (pop_max - pop_med)
    
    if(domc < 0):
        domc = 0
    
    return domc
```

```python
def rateElevation(city, elevation):
    doml = 0
    domh = 0
    dome = 0
    
    low_elevation_data = onto.search(is_a = onto.LoPoint, isLowestPointOf = city.isCityOf[0])
    high_elevation_data = onto.search(is_a = onto.HiPoint, isHighestPointOf = city.isCityOf[0])
    
    lo_elev = int(low_elevation_data[0].loElevation[0])
    hi_elev = int(high_elevation_data[0].hiElevation[0])

    if(elevation == "Low" or elevation == "low"):
        doml = (le_max - lo_elev) / (le_max - le_min)
        domh = (he_max - hi_elev) / (he_max - he_min)        
           
    elif(elevation == "High" or elevation == "high"):
        doml = (lo_elev - le_min) / (le_max - le_min)
        domh = (hi_elev - he_min) / (he_max - he_min)
    
    dome = (doml + domh)/2
    
    return dome
```

```python
def recommendation_rating(result, stateDense, cityPop, elevation, importantFeat):
        
    values = [["stateDense", rateState(result, stateDense)], ["cityPop", rateCity(result, cityPop)], ["elevation", rateElevation(result, elevation)]]
    weights = [1/3, 1/3, 1/3]
    
    max_value = 0
    for j in range(len(values)):
        if(values[j][0] > values[max_value][0]):
            max_value = j
            
    if(values[max_value][1]<0.5):
        rec_desc = "-"
    elif(values[max_value][0]=="stateDense"):
        rec_desc = stateDense + " Population Density"
    elif(values[max_value][0]=="cityPop"):
        rec_desc = cityPop + " Population"
    elif(values[max_value][0]=="elevation"):
        rec_desc = elevation + " Elevation"         
    
    if(importantFeat == "Population" or importantFeat == "population"):
        weights[:] = [2/4, 1/4, 1/4]
        values.insert(0, values.pop(1))

    elif(importantFeat == "Density" or importantFeat == "density"):
        weights[:] = [2/4, 1/4, 1/4]
        values.insert(0, values.pop(0))
    elif(importantFeat == "Elevation" or importantFeat == "elevation"):
        weights[:] = [2/4, 1/4, 1/4]
        values.insert(0, values.pop(2))
    else:
        values = sorted(values, key=lambda l:l[1])
        
#     print(values)    
 
    rec_rate = 0
   
    for i in range(len(values)):
        rec_rate = rec_rate + (values[i][1] * weights[i])
    
    return rec_rate, rec_desc
```

```python
def processInput(city, state, stateDense, cityPop, elevation, importantFeat):
    data = {'city': [],
            'population': [],
            'state': [],
            'state_density': [],
            'lowest_elevation': [], 
            'highest_elevation': [],
            'capital': [],
            'rec_rating': [],
            'description': []}

    df_rec = pd.DataFrame(data)

#     results = onto.search(is_a = onto.City, isCityOf = onto.search(iri=state, type = onto.State))
    city_results = onto.search(iri=city, type = onto.City)
    state_results = onto.search(iri=state, type = onto.State)
    
    results = []
    
    for city in city_results:
        for state in state_results:
            if(city.isCityOf[0]==state):
                results.append(city)
                break
            
#     print(results)
    
    for result in results:
        low_results = onto.search(type = onto.LoPoint, isLowestPointOf = result.isCityOf[0])
        high_results = onto.search(type = onto.HiPoint, isHighestPointOf = result.isCityOf[0])

        if(result.__class__.name == "Capital"):
            capital = True
        else:
            capital = False
        
        pop_dense = int(result.isCityOf[0].stateArea[0]) / int(result.isCityOf[0].statePopulation[0])
        
        lowest_el = low_results[0].loElevation[0]
        highest_el = high_results[0].hiElevation[0]
        
        rec_rating, rec_desc = recommendation_rating(result, stateDense, cityPop, elevation, importantFeat)
        
       
        result_data = {'city': result.name[:-2], 
                       'population': int(result.cityPopulation[0]), 
                       'state': result.isCityOf[0].name, 
                       'state_density': pop_dense,
                       'lowest_elevation': lowest_el, 
                       'highest_elevation': highest_el,
                       'capital': capital,
                       'rec_rating': rec_rating, 
                       'description': rec_desc}
        df_rec = df_rec.append(result_data, ignore_index=True)
        df_rec = df_rec.sort_values('rec_rating', ascending = False)
    
    return df_rec
```

```python
def main():
    print("==== Ontology-Based Fuzzy Logic Recommender Systems ====")
    print()
    city = "*" + input("Masukkan city yang diinginkan\t\t\t\t: ") + "*"
    state =  "*" + input("Masukkan state yang diinginkan\t\t\t\t: ") + "*"
    stateDense = input("Populasi density (Urban, Suburban, Rural)\t\t: ")
    cityPop = input("Populasi (Small, Medium, Large)\t\t\t\t: ")
    elevation = input("Elevasi (Low, High)\t\t\t\t\t: ")
    importantFeat = input("Important Feature (Population, Density, Elevation)\t: ")
    print()
    print("========================================================")
    
    df = processInput(city, state, stateDense, cityPop, elevation, importantFeat)
    df.index += 1 
    with pd.option_context('display.max_columns', None,
                           'display.width', 1000,
                           'display.colheader_justify', 'center'):
#         display(df[:10])
        df = df[["city", "rec_rating"]]
        display(df[:5])
```
